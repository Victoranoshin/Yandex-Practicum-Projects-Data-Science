# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, то ему можно предложить промокоды и специальные условия. Собраны персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

## Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).

Также доступны такие услуги:
- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Цель 
Выбрать модель, которая будет давать максимальное значение метрики ROC-AUC для предсказания, собирается ли клиент уйти.

### План работы
1. Загрузка данных, первичное знакомство с данными:
    - Загрузка данных;
    - Получение информации о типах значений в столбцах, размере датасета;
    - Поиск дубликатов;
    - Поиск уникальных значений;
    - Изучение связи `customerID` в разных датасетах.
2. Предобработка данных:
    - Изменение типов данных в столбцах;
    - Замена значений в столбцах;
    - Заполнение пропусков.
2. Исследовательский анализ данных:
    - Изучение корреляции столбцов в матрице и удаление столбцов, которые сильно коррелируют;
    - Построение гистограммы для численных признаков;
    - Изучение распределение численных признаков и числа клиентов по различным категориям.
3. Моделирование
    - Выделение target и features;
    - Разделение на обучающую, валидационную и тестовую выборки;
    - Кодирование категориальных значений;
    - Обучение и подбор гиперпараметров для моделей LogisticRegression, RandomForestClassifier, LGBMClassifier, CatBoostClassifier.
4. Тестирование и выводы
    - Тестирование выбранной модели (на тестовых данных);
    - Проверка адекватности;
    - Выводы.

## Предобработка данных

```python
data_contact.loc[data_contact['end_date']!='No', 'end_date'] = True 
```

```python
data_contact = data_contact.replace(
    {
        'end_date' : {'No' : False},
        'paperless_billing' : {'Yes' : True, 'No' : False}, 
        'total_charges' : {' ' : None}
    }
)
data_contact = data_contact.astype(
    {
        'type' : 'category',
        'payment_method' : 'category',
        'total_charges' : 'float64'
    }
)
data_contact['begin_date'] = pd.to_datetime(data_contact['begin_date'], format='%Y-%m-%d')
```

```python
data_personal = data_personal.replace(
    {
        'partner' : {'Yes': True, 'No': False}, 
        'dependents' : {'Yes' : True, 'No': False},
        'gender' : {'Male' : True, 'Female' : False}
    }
)
data_personal=data_personal.rename(columns={'gender' : 'is_male'})
data_personal = data_personal.astype({'senior_citizen' : 'bool'})
```

```python
yes_no_set = {'Yes': True, 'No': False} 
data_internet = data_internet.replace(
    {'online_security' : yes_no_set,
     'online_backup' : yes_no_set, 
     'device_protection' : yes_no_set,
     'tech_support' : yes_no_set,
     'streaming_tv' : yes_no_set,
     'streaming_movies' : yes_no_set
    }
)
```

```python
data_phone = data_phone.replace({'multiple_lines':{'Yes': True, 'No': False}})
```

```python
data_all = data_contact.merge(data_personal, how='left', left_on='customer_id', right_on='customer_id').\
merge(data_internet, how='left', left_on='customer_id', right_on='customer_id').\
merge(data_phone, how ='left', left_on='customer_id', right_on='customer_id')
```

Предпологаем, что пропущенные значения в столбце `multiple_lines` означают, что человек не пользуется многоканальностью.

```python
indexes = data_all.loc[data_all['total_charges'].isna(), 'total_charges'].index
data_all.loc[indexes, 'total_charges'] = data_all.loc[indexes, 'monthly_charges']
```

```python
data_all = data_all.fillna(
    {'multiple_lines' : False, 'streaming_movies' : False, 'streaming_tv' : False, 
     'tech_support' : False,'device_protection' : False, 'online_backup' : False,
     'online_security' : False,'internet_service' : 'No'
    }
)
```

Значения количества клиентов, которые ушли и средней общей суммы для этих людей почти во всех случаях меньше для разных категорий в столбцах `type`, `payment_method`, `internet_service`. Только величина средней суммы оплаты для типов оплаты *One year*, *Two year* больше для людей, которые ушли. 

```python
dict_name_kinds = { 
    'is_male' : ['мужчины', 'женщины'],
    'paperless_billing' : ['безналичный расчет', 'наличный наличный'], 
    'senior_citizen' : ['пенсионер(ка)', 'не пенсионер(ка)'],
    'partner' : ['есть супруга(и)', 'нет супруга(и)'],
    'dependents' : ['есть иждивенцы', 'нет иждивенцев'],
    'online_security' : ['есть опция онлайн безопасности', 'нет опции онлайн безопасности'],
    'online_backup' : ['есть опция онлайн поддержки', 'нет опции онлайн поддержки'],
    'device_protection' : ['есть опция защиты устройства', 'нет опции защиты устройства'],
    'tech_support' : ['есть опция техподдержки', 'нет опции техподдержки'],
    'streaming_tv' : ['есть опция интернет-телевидения', 'нет опции интернет-телевидения'],
    'multiple_lines' : ['несколько линий', 'одна линия']
}
```

```python

fig = plt.figure(figsize=(30,30))
fig.suptitle('Отношение числа людей в разных категориях', size=40)
i = 0
offset=0.4
plt.xticks([])
plt.yticks([])

color_palettes = sns.color_palette('bright')[3:5]

plt.legend(handles= [mpatches.Patch(color=color_palettes[0], label='Клиент c нами'),
                    mpatches.Patch(color=color_palettes[1], label='Клиент ушел')], loc='lower right', prop={'size':30}) 

for column_name, frac_names in dict_name_kinds.items(): 
    
    sns.set_theme(font="arial", font_scale=2, style='dark')
    
    data_value_counts = new_dat = data_all.groupby([column_name, 'end_date']).count()['total_charges'] 

    labels = [frac_names[0] if k else frac_names[1] for k in data_value_counts.index[1:3][0]]

    colors = sns.color_palette('bright')[1:3]
    
    fig.add_subplot(4, 3, i+1)
    
    plt.pie(x=data_value_counts.groupby(column_name).sum(), labels=labels, colors=colors, autopct='%.0f%%', 
            wedgeprops=dict(width=offset, edgecolor='w'), radius=1, pctdistance=0.8) 
    
    plt.pie(data_value_counts, radius=1-offset, colors=color_palettes, wedgeprops=dict(width=offset, edgecolor='w'))
    i += 1
```

    1. Больше всего средняя общая сумма оплаты услуг при оплате типа *two year*. Самая маленькая величина средней общей суммы оплаты при оплате *month-to-month*. Больше всего людей оплачивают услуги *month-to-month*, меньше всего людей выбирают оплату *one year*.
    2. Средняя сумма оплаты мнимальна для метода *mailed check*, а максимальна для способов *credit card (automatic)* и *bank transfer (automatic)*. Больше всего клиентов пользуются методом оплаты *electronic check*, меньше всего - *credit card (automatic)*.
    3. Средняя сумма оплаты меньше всего для людей, которые не пользуются услугой интернет-провайдера, а больше всего для категории провайдера *Fiber optic*. Меньше всего клиентов не пользуются услугой интернет-провайдера, а наибольшее число клиентов пользуются категорией провайдера *Fiber optic*.
    
    4. Значения количества клиентов, которые ушли и средней общей суммы для этих людей почти во всех случаях меньше для разных категорий в столбцах `type`, `payment_method`, `internet_service`. Только величина средней суммы оплаты для типов оплаты *One year*, *Two year* больше для людей, которые ушли. 

3. 
    1. Столбцы, для которых оношение числа клиентов в разных категориях (без учета, ушел ли клиент) примерно равно 1 к 1: `is_male`, `partner`.
    2. Столбцы, для которых оношение числа клиентов в разных категориях (без учета, ушел ли клиент) примерно равно 3 к 2: `paperless_billing`, `streaming_tv`, `multiple_lines`.
    3. Столбцы, для которых оношение числа клиентов в разных категориях (без учета, ушел ли клиент) примерно равно 7 к 3: `dependents`, `online_security`, `online_backup`, `device_protection`, `tech_support`.
    4. Отношение числа людей, которые не достигли пенсионного возраста к числу людей, ставших пенсионерами примерно равно 84 к 16.

## Моделирование

### Подготовка выборок

```python
features = data_all.drop(['end_date'], axis=1)
target = data_all['end_date']
```

```python
RS = 130323
```

```python
features_train_valid, features_test, target_train_valid, target_test = train_test_split(
    features, target, test_size=0.25, random_state=RS, stratify=target)
```

```python
features_train_valid.shape
```

```python
scaler = StandardScaler() 
features_train_valid['total_charges'] = scaler.fit_transform(features_train_valid[['total_charges']])
features_test['total_charges'] = scaler.transform(features_test[['total_charges']])
```

```python
code_columns = ['type', 'payment_method', 'internet_service'] 
ohe = OneHotEncoder(drop = 'first') 
ohe.fit(features_train_valid[code_columns])

array_hot_encoded_train_valid = ohe.transform(features_train_valid[code_columns]).toarray()
array_hot_encoded_test = ohe.transform(features_test[code_columns]).toarray()

data_hot_encoded_train_valid = pd.DataFrame(array_hot_encoded_train_valid, index=features_train_valid.index)
data_hot_encoded_test = pd.DataFrame(array_hot_encoded_test, index=features_test.index)

data_other_cols_train_valid = features_train_valid.drop(columns=code_columns)
data_other_cols_test = features_test.drop(columns=code_columns)

train_valid_features_hot_encoded = pd.concat([data_hot_encoded_train_valid, data_other_cols_train_valid], axis=1)
test_features_hot_encoded = pd.concat([data_hot_encoded_test, data_other_cols_test], axis=1)
```

```python
train_valid_features_hot_encoded.columns = train_valid_features_hot_encoded.columns.astype(str) 
```

```python
train_valid_features_hot_encoded.head()
```

```python
le = LabelEncoder()

features_train_valid_labeled = features_train_valid.copy()
features_test_labeled = features_test.copy()

for col in code_columns:
    le.fit(features_train_valid[col])
    features_train_valid_labeled[col] = le.transform(features_train_valid[col]) 
    features_test_labeled[col] = le.transform(features_test[col])
```

```python
features_train_valid_labeled.head()
```

Разделили данные на выборки, сделали масштабирование признаков, создали выборки с категориальными значениями, закодированными с помощью LabelEncoder и OHE отдельно.

### Случайный лес

```python
sklearn_weights = compute_class_weight(class_weight='balanced',
                                                    y=target_train_valid,classes=np.unique(target_train_valid))
weights_dict = dict(zip([False, True], sklearn_weights)) 
weights_dict
```

```python
imba_pipeline = make_pipeline(
                              RandomForestClassifier(random_state=RS, 
                                                     class_weight=weights_dict))
params = {'max_depth':range(5, 10, 1), 'n_estimators':range(195, 210, 1)}
new_params = {'randomforestclassifier__' + key: params[key] for key in params}

grid_imba = GridSearchCV(imba_pipeline, param_grid=new_params, cv=
                         StratifiedKFold(4, shuffle=True, random_state=RS), scoring='roc_auc', n_jobs=-1) 
```

```python
%%time
grid_imba.fit(features_train_valid_labeled, target_train_valid)
```

```python
best_forest = grid_imba.best_estimator_.named_steps['randomforestclassifier']
best_forest
```

```python
score_forest = grid_imba.best_score_
print('Значение roc_auc для лучших параметров случайного леса на валидационных данных:', score_forest)
```

### Логистическая регрессия


```python
imba_pipeline = make_pipeline(
                              LogisticRegression(max_iter=1000, random_state=RS,  class_weight=weights_dict)) 
                                                     
params = {'solver':['lbfgs', 'liblinear']}
new_params = {'logisticregression__' + key: params[key] for key in params}

grid_imba = GridSearchCV(imba_pipeline, param_grid=new_params, cv=
                         StratifiedKFold(4, shuffle=True, random_state=RS), scoring='roc_auc', n_jobs=-1)
```

```python
%%time
grid_imba.fit(train_valid_features_hot_encoded, target_train_valid)
```

```python
best_lr = grid_imba.best_estimator_.named_steps['logisticregression']
best_lr
```

```python
score_lr = grid_imba.best_score_
print('Значение roc_auc для лучших параметров линейной регрессии на валидационных данных:', score_lr)
```

### LightGBM

```python
imba_pipeline = make_pipeline(
                              LGBMClassifier(random_state=RS,  
                                                     class_weight=weights_dict))
params = {'max_depth':range(2, 8, 1), 'n_estimators':range(40, 55, 1) }
new_params = {'lgbmclassifier__' + key: params[key] for key in params}

grid_imba = GridSearchCV(imba_pipeline, param_grid=new_params, cv=
                         StratifiedKFold(4, shuffle=True, random_state=RS), scoring='roc_auc', n_jobs=-1)
```

```python
%%time
grid_imba.fit(features_train_valid_labeled, target_train_valid)
```

```python
score_LGBM = grid_imba.best_score_
print('Значение roc_auc для лучших параметров LightGBM на валидационных данных:', score_LGBM)
```

### CatBoost

```python
imba_pipeline = make_pipeline(
                              CatBoostClassifier(random_state=RS))#, class_weight=weights_dict
params = {'learning_rate': [0.01, 0.003, 0.002, 0.001], 'depth':[4, 5, 6, 7]}
new_params = {'catboostclassifier__' + key: params[key] for key in params}

grid_imba = GridSearchCV(imba_pipeline, param_grid=new_params, cv=
                         StratifiedKFold(4, shuffle=True, random_state=RS), scoring='roc_auc', n_jobs=-1,
                          error_score='raise')
```

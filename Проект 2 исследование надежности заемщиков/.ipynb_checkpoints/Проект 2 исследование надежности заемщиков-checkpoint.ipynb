## Предобработка данных

### Удаление пропусков 1

```python
data.isna().sum()
```

В двух столбцах `days_employed`, `total_income` есть пропущенные значения. На сумму дохода больше всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. 

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Обработка аномальных значений

Заменим все отрицательные значения в столбце `days_employed` положительными с помощью метода.

```python
data['days_employed'] = data['days_employed'].abs()
```

Для каждого типа занятости выведим медианное значение трудового стажа `days_employed` в днях.

```python
data.groupby('income_type')['days_employed'].agg('median')
```

У двух типов (безработные и пенсионеры) получатся аномально большие значения, сложно определить причины их появления, данный стобец не понадобится в нашем исследовании, поэтому не будем его менять.

```python
data['children'].unique()#  перечень уникальных значений столбца `children`.
```

Удалим строки, в которых аномальные значения из датафрейма `data`.

```python
data = data[(data['children'] != -1) & (data['children'] != 20)]
```

```python
data['children'].unique()
```

### Удаление пропусков 2

Заполним пропуски в столбце `days_employed` медианными значениями по каждого типа занятости `income_type`.

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == t), 'days_employed'].median()
```

```python
data.isna().sum()
```

### Категоризация данных

Зададим в датафрейме `data` столбец `total_income_category` с категориями дохода:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.



```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

```python
data['purpose'].unique() # перечень уникальных целей взятия кредита
```

Cформируем новый столбец `purpose_category`, в который войдут следующие категории:

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## Исследуем данные

Проверим, есть ли зависимость между количеством детей и возвратом кредита в срок.

```python
def pivot_count_sum_mean_debt_function(index_name ):# функция возвращает сводную таблицу с информацией о должниках с группировкой по index_name
    data_pivot = data.pivot_table(index=index_name, values='debt', aggfunc={'count', 'sum', 'mean'})# создание сводной таблицы
    data_pivot = data_pivot.rename(columns={"count":"Всего кредитополучателей", "sum": "Всего должников", "mean":"Доля должников"}) #переименование столбцов
    data_pivot = data_pivot.sort_values(by='Доля должников', ascending=False)# сортировка по убыванию доли должников
    data_pivot = data_pivot.reindex(columns=['Всего кредитополучателей', 'Всего должников', 'Доля должников']) #изменение расположения колонок
    return data_pivot
```

```python
pivot_count_sum_mean_debt_function('children') # получение информации о должниках с группировкой по количеству детей
```

**Вывод:**

Больше всего доля людей с долгами среди родителей четырех детей. Имеется информация только о девяти людях с пятью детьми, поэтому невозможно сделать корректные выводы для этой категории.

Доля людей с долгами возрастает при увеличении количества детей от нуля до двух. Доля людей с долгами с тремя детьми меньше, чем людей с одним ребенком, но больше, чем у людей без детей.

Проверим, есть ли зависимость между семейным положением и возвратом кредита в срок.

```python
pivot_count_sum_mean_debt_function('family_status')  # получение информации о должниках с группировкой по семейному положению
```

**Вывод:** 
Меньше всего доля людей с долгами среди людей, лишившихся мужа/жены. Больше всего доля людей с долгами среди не вступавших в брак. 

Среди женатых/замужних доля людей с долгами больше, чем среди людей в разводе, но меньше, чем среди людей в гражданском браке.

Проверим, существует ли зависимость между уровнем дохода и возвратом кредита в срок.

```python
pivot_count_sum_mean_debt_function('total_income_category') # получение информации о должниках с группировкой по уровню дохода
```

**Вывод:** 

Кредитополучателей категории "B" и "С" гораздо больше, чем в других категориях в предоставленной таблице.

При увеличении дохода доля должников уменьшается более чем на 1%, если рассматривать только категории "B" и "C". 
По остальным категориям четкой зависимости между уровнем дохода и долей должников обнаружить нельзя.

Выясним, влияют ли разные цели кредита влияют на его возврат в срок.

```python
pivot_count_sum_mean_debt_function('purpose_category') # получение информации о должниках с группировкой по цели взятия кредита
```

**Вывод:** 

Меньше всего доля должников среди людей, берущих кредит с целью проведения операций с недвижимостью. 
Больше всего доля должников среди людей, у которых цель взятия кредита - проведение операций с автомобилем. 

Доля должников, берущих кредит для получения образования немного меньше, чем среди людей, которым нужен кредит для проведения операций с автомобилем. Доля людей с долгами, берущих кредит для проведения свадьбы примерно на 1 % больше чем среди людей, которые берут кредит для проведения операций с недвижимостью. 

**Возмможные причины появления пропусков в исходных данных:**

Пропуски в данных были в полях `"total_income"`, `"days_employed"`. Возможно, при вводе данных сотрудник банка совершил ошибку или произошел сбой в системе. Также возможно часть людей в приведенных данных являются сотрудниками банка, и информация о доходе и количестве рабочих дней уже имеется в отдельной таблице, данные из которой не были перенесены в представленную таблицу.
